```python
import os
os.system("pip install pymupdf")

import fitz  # PyMuPDF
import matplotlib.pyplot as plt
import numpy as np

# Open the PDF file
pdf_path = "./output/demo1/hybrid_auto/demo1_layout.pdf"
pdf_doc = fitz.open(pdf_path)
# Load and render the first page
layout_page_0 = pdf_doc.load_page(0)
layout_page_1 = pdf_doc.load_page(1)

pix1 = layout_page_0.get_pixmap()
img_array1 = np.frombuffer(pix1.samples, dtype=np.uint8).reshape(pix1.height, pix1.width, pix1.n)

# Load and render the second page
pix2 = layout_page_1.get_pixmap()
img_array2 = np.frombuffer(pix2.samples, dtype=np.uint8).reshape(pix2.height, pix2.width, pix2.n)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 8))

# Display the first image
axes[0].imshow(img_array1)
axes[0].axis('off')

# Display the second image
axes[1].imshow(img_array2)
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

#### **Convert multiple PDFs in the same directory**

```python
!mineru -p ./ -o ./output -b pipeline
```

```python
import fitz
from IPython.display import display, Image as IPImage
from PIL import Image as PILImage
import io

def render_pdf_first_page(pdf_path, zoom=2):
    doc = fitz.open(pdf_path)
    page = doc.load_page(0)
    mat = fitz.Matrix(zoom, zoom)
    pix = page.get_pixmap(matrix=mat)
    img_data = pix.tobytes("png")
    doc.close()
    return PILImage.open(io.BytesIO(img_data))

# 加载两个 PDF 第一页
img1 = render_pdf_first_page("output/demo1/auto/demo1_layout.pdf")
img2 = render_pdf_first_page("output/demo2/auto/demo2_layout.pdf")

# 创建一个新图像：左右拼接
total_width = img1.width + img2.width
max_height = max(img1.height, img2.height)

combined_img = PILImage.new('RGB', (total_width, max_height))
combined_img.paste(img1, (0, 0))
combined_img.paste(img2, (img1.width, 0))

# 将图像转为 BytesIO，用于 IPython 显示
img_byte_arr = io.BytesIO()
combined_img.save(img_byte_arr, format='PNG')
final_image = IPImage(data=img_byte_arr.getvalue())

# 显示最终左右拼接的图像
print("左右拼接的两个 PDF 首页：")
display(final_image)
```
